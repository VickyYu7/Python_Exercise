```python
import csv
with open('names.csv') as fn:
    names = list(csv.reader(fn))
names
customer = {n[0]: n[1] for n in names}
customer
```

```python
with open('items.csv') as fi:
    stuff = list(csv.reader(fi))

# purchase = {i[0]: i[1] for i in stuff} 
# old items are overlapped by new items, 一个id买了不同的item，只有new item会被记录
purchase = {}

for id_, item in stuff:
    if id_ not in purchase:
        purchase[id_] = [item]
    else:
        purchase[id_].append(item)

purchase
```

```python
result = []
for idt, item in purchase.items():
    n = customer[idt]
    if n:
        result.append((n, item))
result
```

```python
with open('names.csv') as fp:
    names = list(csv.reader(fp))

with open('items.csv') as fn:
    items = list(csv.reader(fn))
```

```python
names = dict(names)
items_by_name= {}
for id_, item in items:
    name = names.get(id_) # no need to do error handling, no IndexError
    if name in items_by_name: # if no key check, error rise in the next line
        items_by_name[name].append(item)
    else:
        items_by_name[name] = [item]
items_by_name
```
